x ve y ayarlama:

```python
y=dataFrame["Type"].values 
x=dataFrame.drop("Type",axis=1).values #ayarlandi ve diziye cevrildi
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=15)
```

kucultme islemleri: 

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
```

```python
scaler.fit(x_train)
```

```python
x_train=scaler.transform(x_train)
```

butun x_train verileri (0,1) arasinda gore transform edilir.

```python
x_test=scaler.transform(x_test)
```

butun x_test verileri (0,1) arasinda gore transform edilir.

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Activation,Dropout
from tensorflow.keras.callbacks import EarlyStopping
```

EarlyStopping,Dropout özellikle overfitting(Aşırı öğrenme) problemlerini cozmek icin kullaniliriz.
Aşırı öğrenme: üretilen bir analizin belirli bir veri kümesine aşırı oranda uyum göstermesi, dolayısıyla bu veri kümesinde yer almayan yeni verilere uyum sağlayamaması problemidir. 

```python
x_train.shape
```

```python
model=Sequential()
#units=30 iceriye 30 tane noron koy.Peki neden 30 => kac tane kolonun varsa input layeri buna esit yapilmasi onerilir.
model.add(Dense(units=30,activation="relu"))
model.add(Dense(units=15,activation="relu"))
model.add(Dense(units=15,activation="relu"))
model.add(Dense(units=1,activation="sigmoid"))  #output layerina sigmoid aktivasyon fonksiyonunu verdik

model.compile(loss="binary_crossentropy",optimizer="adam")
#siniflandirma olarak binary bir islem yaptigimiz icin kayip fonksiyonun (loss) binary_crossentropy seceriz.
```

train etme:

```python
earlyStopping=EarlyStopping(monitor="val_loss",mode="min",verbose=1,patience=25)
```

```python
#model.fit(x=x_train,y=y_train,epochs=700,validation_data=(x_test,y_test),verbose=1)
#model overfittinge ugrar
model.fit(x=x_train,y=y_train,epochs=700,validation_data=(x_test,y_test),verbose=1,callbacks=[earlyStopping])
```

```python
modelKaybi=pd.DataFrame(model.history.history)
```

```python
modelKaybi.plot()
```

Dropout:

```python
model_2=Sequential()

model_2.add(Dense(units=30,activation="relu"))
model_2.add(Dropout(0.6))
model_2.add(Dense(units=15,activation="relu"))
model_2.add(Dropout(0.6))
model_2.add(Dense(units=15,activation="relu"))
model_2.add(Dropout(0.6))
model_2.add(Dense(units=1,activation="sigmoid"))  

model_2.compile(loss="binary_crossentropy",optimizer="adam")
```

```python
model_2.fit(x=x_train,y=y_train,epochs=700,validation_data=(x_test,y_test),verbose=1,callbacks=[earlyStopping])
```

```python
kayipDf=pd.DataFrame(model_2.history.history)
```

```python
kayipDf.plot()
```

```python
#siniflari tahmin etme. Verdigin girdilere gore siniflarini tahmin etme:(predict_classes desteklenmiyor)
tahminlerimiz=model.predict(x_test)
tahminlerimiz
```

```python
esik_degeri = 0.5
tahminler = (tahminlerimiz > esik_degeri).astype(int)
```

```python
tahminler
```

siniflarndirmanin ne kadar duzgun sonuclar verdigini degerlendirme:

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,tahminler))
```

precision:ne kadar duyarli tahmin eder

```python
print(confusion_matrix(y_test,tahminler))
```
